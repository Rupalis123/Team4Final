```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from sklearn import tree
```

```python
# Read the csv file into a pandas DataFrame

covid_df = pd.read_csv('test.csv')
covid_df
```

```python
# Assign the data to X and y
# Note: Sklearn requires a two-dimensional array of values
# so we use reshape to create this

X = covid_df["Mean Age"].values.reshape(-1, 1)
y = covid_df["Mortality"].values.reshape(-1, 1)

print("Shape: ", X.shape, y.shape)
```

```python
# Plot the data to see if a linear trend exists

### BEGIN SOLUTION
plt.scatter(X, y)
plt.xlabel("Mean of Age of Patient")
plt.ylabel("Mortality Rate")
### END SOLUTION
```

```python
# Use sklearn's `train_test_split` to split the data into training and testing

from sklearn.model_selection import train_test_split

target = 'Mortality'
X = covid_df.drop(target, axis='columns', inplace=False)
y = covid_df[target]

X_unselected, X_selected, y_unselected, y_selected = train_test_split(X, y, stratify=y, random_state=1, test_size=100)
```
