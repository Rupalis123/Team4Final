```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
from sklearn import tree
```

```python
df = pd.read_csv('covid_clinical_data.csv')
```

```python
my_list = df.columns.values.tolist()
```

```python
selected = df.iloc[:,4:]
selected
```

```python
df = selected.fillna(0.0)
df.head()
```

```python
my_list
```

```python
dropped_df = df.drop(['Obesity', 'Any Comorbidity','temp','Loss of smell/taste','Asymptomatic','Chest Pain','Upper air-way congestion','Respiratory rate > 24 breaths per min','Cerebrovascular Disease'], axis=1)
```

```python
# master df
dropped_df.rename(columns={"Fever (temperature ‚â•37¬∑3¬∞C)": "Fever", "Diarrhoea": "Diarrhea"})
```

```python
#age risk factor using linear regression 
```

```python
target=dropped_df['Mortality']
target_names=['risk','not_risk']
```

```python
data = dropped_df.drop("Mortality", axis=1)
feature_names = data.columns
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=1)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)  
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=500)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```

```python
feature_names = dropped_age.columns
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dropped_age, target, random_state=42)
```
